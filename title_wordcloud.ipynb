## 스타일별로 워드 클라우드 그리기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud
```

```python
# 키워드 - 빈도 데이터프레임 import
neurope = pd.read_csv('neurope.csv')
vintage = pd.read_csv('vintage.csv')
classic = pd.read_csv('classic.csv')
french = pd.read_csv('french.csv')
lovely = pd.read_csv('lovely.csv')
industry = pd.read_csv('industry.csv')
asia = pd.read_csv('asia.csv')
```

```python
# 불용어(자주 등장하지만 의미없는 단어)
STOPWORDS = {'아파트', '인테리어', '공간', '있는', '신혼집', '주택', '하우스', '리모델링', '취향', '부부', '구축', '시공', '생활', '전원', '같은', '겔러리', '대면', '행복한', '주방', '같은', '셀프', '자취', '화이트', '디자이너', '작은', '일링', '원룸', '스타일링', '우리', '좋아', '가구', '우드', '완성', '빌라', '무드', '소품', ' 채운', '시간','더해진', '주택', '가득'}
```

```python
icon1 = Image.open('home.png')
plt.imshow(icon1)

mask1 = Image.new("RGB", icon1.size, (255,255,255))
mask1.paste(icon1,icon1)
mask1 = np.array(mask1)
```

```python
icon2 = Image.open('desk-lamp.png')
plt.imshow(icon2)

mask2 = Image.new("RGB", icon2.size, (255,255,255))
mask2.paste(icon2,icon2)
mask2 = np.array(mask2)
```

```python
icon3 = Image.open('sofa.png')
plt.imshow(icon3)

mask3 = Image.new("RGB", icon3.size, (255,255,255))
mask3.paste(icon3,icon3)
mask3 = np.array(mask3)
```

```python
icon4 = Image.open('chair.png')
plt.imshow(icon4)

mask4 = Image.new("RGB", icon4.size, (255,255,255))
mask4.paste(icon4,icon4)
mask4 = np.array(mask4)
```

-  빈티지&레트로 = ['빈티지', '레트로', '원목', '브릭']

```python
count_dic = dict(zip(vintage.단어, vintage.빈도))
```

```python
from numpy import newaxis
for i in STOPWORDS:
    if i in count_dic:
        del count_dic[i]
```

```python
wc = WordCloud(
font_path='잘풀리는하루 Medium.ttf', #폰트
background_color = 'white', #배경색
max_words = 100, #단어개수
prefer_horizontal = 1, 
width = 400,height = 300,
colormap = 'vintage2',
mask = mask2)

cloud = wc.generate_from_frequencies(count_dic)

plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(cloud,interpolation="bilinear")
```
