```python
prd_re['별점'].mode() #최빈값 5.0 

# 결론 사람들은 대부분 별점을 5.0을 준다.
#실제로 만족도가 높아서일까?
```

```python
prd_re['별점'].mean() #평균 별점은 4.7이군
```

```python
prd_re['별점'].median() #중간값
```

```python
prd_re['별점'].std() 
#표준편차 0.5 그렇다면 별점을 줄때 5.0 / 4.5/ 4.0/ 3.5 이렇게 주는걸로 추측
```

```python
import pingouin as pg
import numpy as np
```

```python
import scipy
scipy.stats.bootstrap([prd_re], np.mean)
```

```python
pg.ttest(prd_re['별점'] , 0, confidence=0.95)
```

```python
pg.ttest(prd_re['별점'], 5.0, confidence=0.95)
#별점이 5.0이 대부분이다?
```
