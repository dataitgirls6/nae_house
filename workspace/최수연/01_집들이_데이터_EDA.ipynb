### ① 스타일 '내추럴'의 title 데이터

```python
import pandas as pd
df = pd.read_csv("housewarming2.csv")
df_natural = df['제목'][df['스타일']=='내추럴'].to_list()
df_natural
```

```python
title_natural = ''
for i in df_natural:
    i.replace('\n','').replace('.','')

    title_natural = title_natural + i

print(title_natural)
```

### ② title 데이터 형태소 분석

```python
# 필요 묘듈 임포트
import pandas as pd
from konlpy.tag import Okt
import re

# 파일 로드 
# df_natural 파일
# 데이터 전처리
okt = Okt()
doc_ko = re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s]", "", title_natural)
token_ko = pd.DataFrame(okt.pos(doc_ko))
token_ko.columns = ['단어', '형태소']

# print(token_ko)
tt_group = token_ko.groupby('단어')['형태소'].count()
tt_group.sort_values(ascending=False)

#result = token_ko.value_counts().reset_index(name='count')

# a = result[:][(result['단어'].apply(lambda x:len(x) >= 2)) & (result['형태소'] == 'Noun') | (result['형태소'] == 'Adverb')].head(500)
# a.to_csv("수사기관사칭형_통합.csv")
```
