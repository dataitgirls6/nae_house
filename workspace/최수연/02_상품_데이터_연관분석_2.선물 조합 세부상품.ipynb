```python
import pandas as pd

prd_comb = pd.read_excel('prd_comb.xlsx')
prd_comb.head()
```

```python
# 데이터 로드(생성)
data = list(prd_comb['상품조합_edit'])
```

```python
# 2. 데이터 생성
data_comb = []

for i in data:
    data_comb.append(i.split(','))

print(data_comb)
```

```python
# 3. 필요한 패키지 로드
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder   # 연관분석을 위한 데이터 전처리 모듈
from mlxtend.frequent_patterns import apriori          # 연관분석 모듈
```

```python
# 4. 데이터 전처리
te = TransactionEncoder()        # 연관분석을 하기 위해 데이터 전처리 클래스를 객체화 시킨다
te.fit(data_comb)

te_ary = te.transform(data_comb)   # 위에서 계산된 내용으로 데이터를 변경합니다.
te_ary
```

```python
len(te_ary[0])
```

```python
print(te.columns_)
```

```python
len(te.columns_)
```

```python
comb_df = pd.DataFrame(te_ary, columns = te.columns_)
comb_df
```

```python
# 지지도 0.01 이상인것만 출력

apriori(comb_df, min_support = 0.02, use_colnames=True)
```

```python
# 지지도 0.01, 신뢰도와 향상도 출력
a = apriori(comb_df, min_support=0.01, use_colnames=True)

from mlxtend.frequent_patterns import association_rules

association_rules(a, metric='confidence', min_threshold= 0.01)
```

```python
# 가장 연관성이 높은 품목
# 지지도 0.01 이상, 신뢰도 0.1 이상 출력
a = apriori(comb_df, min_support = 0.01, use_colnames = True)

from mlxtend.frequent_patterns import association_rules
a3 = association_rules(a, metric='confidence', min_threshold=0.1)
a3.sort_values(by=['support','confidence'], ascending = False)
```

### ● 연관규칙 주요 용어

1. 지지도(Support) : 전체 거래중 항목 A 와 항목 B 를 동시에 포함하는 거래의 비율
2. 신뢰도(Confidence) : A 상품을 샀을 때 B상품을 살 조건부 확률에 대한 척도
3. 향상도(Lift) : 발견한 규칙이 우연에 의한 발생인지 아닌지를 판단하는 척도
