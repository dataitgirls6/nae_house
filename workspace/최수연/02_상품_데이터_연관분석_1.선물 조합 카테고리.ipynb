```python
import pandas as pd

zip_q = pd.read_excel('zipquestion.xlsx')
zip_q.head()
```

```python
zip_q.columns
```

```python
# 데이터 로드(생성)
data = list(zip_q['Q4.  ‘오늘의집’에서 집들이 선물을 구매한다면 어떤 상품을 구매하실건가요? (택2)'])
```

```python
# 2. 데이터 로드(생성)
dataset = []

for i in data:
    dataset.append(i.split(','))

print(dataset)
```

```python
# 3. 필요한 패키지 로드
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder   # 연관분석을 위한 데이터 전처리 모듈
from mlxtend.frequent_patterns import apriori          # 연관분석 모듈
```

```python
# 4. 데이터 전처리
te = TransactionEncoder()        # 연관분석을 하기 위해 데이터 전처리 클래스를 객체화 시킨다
te.fit(dataset)

te_ary = te.transform(dataset)   # 위에서 계산된 내용으로 데이터를 변경합니다.
te_ary
```

```python
df = pd.DataFrame(te_ary, columns = te.columns_)
df
```

```python
# 지지도 0.3 이상인것만 출력

apriori(df, min_support = 0.3, use_colnames=True)
```

```python
# 위에서 전처리한 데이터를 가지고 연관분석을 하는데 가장 연관성이 높은 품목이 서로 어떻게 되는지 
# 지지도 0.1 이상, 신뢰도 0.1 이상인 품목들을 출력
a = apriori(df, min_support = 0.1, use_colnames = True)

from mlxtend.frequent_patterns import association_rules
association_rules(a, metric='confidence', min_threshold=0.1)
```

### ● leverage
#### 예 > 0번. (조명) ↔ (데코/식물)

- 조명과 데코식물이 얼마나 자주 나타나는지 빈도를 관찰해서 둘 사이의 차이를 계산\
조명과 데코식물이 서로 얼마나 독립적인지를 나타내는 척도.\
이 값이 0 이면 서로 독립.
- 범위 : -1 ~ 1

#### leverage(조명 → 데코/식물) 
##### = support(조명 → 데코/식물) - support(조명) x support(데코/식물)

### ● conviction

#### 예 > (조명 → 데코/식물)

####  : 높은 conviction 값은 선행품목과 연관품목이 서로 매우 의존적임을 의미.

#### conviction (조명→데코/식물) = ( 1 - support(데코/식물) ) / (1 - confidence(조명→데코/식물))

- 위의 식에서 분모가 0 이 되면 conviction 의 값이 무한대(inf) 로 표시합니다.\
  향상도와 비슷하게 아이템들끼리 서로 독립적이면 conviction 값이 1 이됩니다.

- 범위: 0 ~ 무한대(∞)

```python
# 위의 결과를 신뢰도가 높은것부터 출력
a = apriori(df, min_support = 0.1, use_colnames = True)

from mlxtend.frequent_patterns import association_rules
a2 = association_rules(a, metric='confidence', min_threshold=0.1)
a2.sort_values(by=['support','confidence'], ascending = False)
```
