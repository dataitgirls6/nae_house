#### - 집들이 게시글에 태깅 많이 된 상품 300개 중 인기순 상품 300위에 해당하는 상품 개수

```python
prd_popul_rnk = pd.read_csv("prd_popul_rnk.csv")

cnt = 1
pop_prd = 0
for i in Counter(hou_popul_rnk['상품url']).most_common(300):
    if i[0] in list(prd_popul_rnk['상품url']):
        pop_prd += 1
        # cnt += 1
    else:
        pass

print(pop_prd)
```

```python
# 판매순 상품 300위 안에 있는 태깅 상품
cnt = 1
pop_rnk = []
for i in Counter(hou_popul_rnk['상품url']).most_common(300):
    if i[0] in list(prd_popul_rnk['상품url']):
        pop_rnk.append(i[0])
        # cnt += 1
    else:
        pass

print(pop_rnk)
```
