```python
df_style_dummies = pd.read_csv('df_style_dummies.csv')
df_style_dummies = df_style_dummies.rename(columns = {'Unnamed: 0' : '상품명'})
df_style_dummies.index = df_style_dummies['상품명']
df_style_dummies = df_style_dummies.drop('상품명', axis = 1)
```

```python
df_style_dummies.head(30)
```

```python
# 상품을 기준으로 분위기의 상관관계 파악
df_style_dummies.corr()
```

- 모던, 미니멀&심플, 내추럴 상관계수가 모두 0.7이상으로 상품 간의 경계가 없을 것
- 상관계수가 0.3이하인 분위기는 클러스터링으로 상품의 분위기를 분류할 수 있지 않을까?

```python
# 프렌치&프로방스, 인더스트리얼, 한국&아시아 클러스터링
from sklearn.cluster import KMeans 
kmeans = KMeans(n_clusters = 4)
kmeans.fit(df_style_dummies[['프렌치&프로방스', '인더스트리얼', '한국&아시아']])
```

```python
labels = kmeans.labels_
df_style_dummies['클러스터'] = labels
df_style_dummies[['프렌치&프로방스', '인더스트리얼', '한국&아시아', '클러스터']]
```

```python
# 클러스터 0 : 한국&아시아
df_style_dummies[df_style_dummies['클러스터'] == 0][['프렌치&프로방스', '인더스트리얼', '한국&아시아', '클러스터']]
```

```python
# 클러스터 1 : 세 가지 분위기에 모두 해당하지 않는 상품들
df_style_dummies[df_style_dummies['클러스터'] == 1][['프렌치&프로방스', '인더스트리얼', '한국&아시아', '클러스터']]
```

```python
# 클러스터 2 : 프렌치&프로방스
df_style_dummies[df_style_dummies['클러스터'] == 2][['프렌치&프로방스', '인더스트리얼', '한국&아시아', '클러스터']]
```
