```python
import pandas as pd
df_hw = pd.read_csv('housewarming2.csv')
df_prd = pd.read_csv('prd_details_fixed.csv')
df_review = pd.read_csv('prd_reviews.csv')
df_style = pd.read_csv('prd_style_match.csv')
df_sales_rnk = pd.read_csv('prd_sales_rnk.csv')
df_popul_rnk = pd.read_csv('prd_popul_rnk.csv')
```

```python
df_prd['대분류'] = df_prd['카테고리'].apply(lambda x : x.split("', ")[0].lstrip("['"))
```

```python
df_prd.columns
```

# 판매량 랭킹 데이터와 상품 데이터 병합

```python
df1 = df_prd[['상품url', '브랜드명', '상품명', '대분류', '상품가격', '리뷰수', '스크랩수', '스타일링샷수']]
df1 = df1.drop_duplicates()
df1
```

```python
df2 = df1.merge(df_sales_rnk[['상품url', '판매랭킹']], how = 'left', on = '상품url')
df2
```

```python
df2 = df2[df2['판매랭킹'].isnull() == False]
df2
```

## 상관분석

```python
df2.corr()
```

## 회귀분석

```python
from statsmodels.formula.api import ols
m = ols('판매랭킹 ~ 스크랩수', data = df2).fit()
m.summary()
```

```python
m = ols('판매랭킹 ~ 리뷰수', data = df2).fit()
m.summary()
```

```python
m = ols('판매랭킹 ~ 스타일링샷수', data = df2).fit()
m.summary()
```

```python
m = ols('판매랭킹 ~ 상품가격', data = df2).fit()
m.summary()
```

## 다중회귀분석

```python
# 두가지 변수
from statsmodels.formula.api import ols
m = ols('판매랭킹 ~ 스크랩수 + 리뷰수', data = df2).fit()
m.summary()
```

```python
m = ols('판매랭킹 ~ 스크랩수 + 스타일링샷수', data = df2).fit()
m.summary()
```

## 상호작용 분석

### 대분류에 따른 절편 변화

```python
from statsmodels.formula.api import ols
ols('판매랭킹 ~ 리뷰수 + 대분류', df2).fit().summary()
```

```python
ols('판매랭킹 ~ 스크랩수 + 대분류', df2).fit().summary()
```

```python
ols('판매랭킹 ~ 스타일링샷수 + 대분류', df2).fit().summary()
```

### 대분류에 따른 기울기 변화

```python
ols('판매랭킹 ~ 리뷰수 + 리뷰수:대분류', df2).fit().summary()
```

```python
ols('판매랭킹 ~ 스크랩수 + 스크랩수:대분류', df2).fit().summary()
```

```python
ols('판매랭킹 ~ 스타일링샷수 + 스타일링샷수:대분류', df2).fit().summary()
```

### 대분류에 따른 기울기와 절편 변화

```python
ols('판매랭킹 ~ 리뷰수 + 대분류 + 리뷰수:대분류', df2).fit().summary()
```

```python
ols('판매랭킹 ~ 스크랩수 + 대분류 + 스크랩수:대분류', df2).fit().summary()
```

```python
ols('판매랭킹 ~ 스타일링샷수 + 대분류 + 스타일링샷수:대분류', df2).fit().summary()
```
