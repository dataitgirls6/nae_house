### 상품 데이터 통계분석  

```python
import pandas as pd
# pd.options.display.max_columns = None
```

```python
df = pd.read_csv('prd_details_fixed.csv')
df2 = pd.read_csv('prd_style_match.csv')
```

```python
df
```

```python
df2
```

```python
df3 = pd.merge(df, df2, how='inner', on='상품명')
df3
```

```python
# pd.set_option('display.max_rows', 500)
tmp = df3.groupby('분위기')['상품명'].value_counts()
tmp
```

```python
prd_count = df3.groupby('상품url', as_index=False)['id'].count()
prd_count
```

```python
master = pd.merge(df3,prd_count, how='inner', on='상품url')
master.drop(columns='id_x', inplace=True)
master.rename(columns = {'id_y': '판매수'},inplace = True)
master
```

```python
master['판매액'] = master['상품가격'] * master['판매수']
master
```

```python
master.dropna(subset='상품가격', inplace=True)
master.drop_duplicates(inplace=True)
master
```

```python
master.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')

plt.rcParams['font.family'] = 'Malgun Gothic'
colormap = plt.cm.PuBu
plt.figure(figsize=(10, 8))
plt.title("Person Correlation of Features", y = 1.05, size = 15)
sns.heatmap(master.corr(), linewidths = 0.1, vmax = 1.0,
           square = True, cmap = colormap, linecolor = "white", annot = True, annot_kws = {"size" : 16})
```

- 많이 팔린 상품은 리뷰수, 스크랩수, 스타일링샷수가 많을까?
    - 약한 ~ 보통의 상관관계가 있다.
- 스크랩수가 많아지면 리뷰수도 많아진다(0.7) -> 구매 확정률이라는 지표

```python
sns.pairplot(master)
plt.show()
```

```python
# 많이 팔린 상품은 분위기가 영향을 미치나? (어떤 분위기를 팔아야 잘 팔릴까?)
# 판매액으로 봐야할지 판매수로 봐야할지 모르겠음
from statsmodels.formula.api import ols
ols('판매수 ~ 분위기', master).fit().summary()
```

- 러블리&로맨틱기준 판매수5, 인더스트리얼+7, 한국&아시아+3 > 프렌치&프로방스+1
    - 인더스트리얼>한국&아시아>프렌치&프로방스>러블리&로맨틱
    - 1위가 인더스트리얼?? 표본의 수가 작아서 이런 결과가?

```python
# 많이 팔린 상품은 분위기가 영향을 미치나? (어떤 분위기를 팔아야 잘 팔릴까?)
from statsmodels.formula.api import ols
ols('판매액 ~ 분위기', master).fit().summary()
```

- 프렌치&프로방스 > T.클래식&앤틱 > 러블리&로맨틱 > 클래식&앤틱
