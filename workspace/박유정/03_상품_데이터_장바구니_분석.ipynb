### 상품 데이터 장바구니 분석

```python
import pandas as pd
```

```python
df = pd.read_csv('prd_details_fixed.csv')
df
```

```python
transaction = df.groupby(['id', '상품명'])['상품명'].count().reset_index(name ='Count')
transaction
```

```python
basket = transaction.pivot_table(index='id', columns='상품명', values='Count', aggfunc='sum').fillna(0)
basket.head()
```

```python
# 0, 1 로 변환해주는 함수
def encoding(x):
    if x<=0:
        return 0
    if x>=1:
        return 1

basket = basket.applymap(encoding)
basket.head()
```

```python
# basket.to_csv('basket.csv', index=False)
```

- '지지도': 전체 항목 중(행들 중) x와 y를 모두 포함하는 경우의 비율

```python
# 기본은 0.5, 0.1부터해야 뭐가 나옴
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
freq_item = apriori(basket, min_support = 0.01,use_colnames = True)
freq_item
```

```python
freq_item.iloc[231:]['itemsets']
```

- 신뢰도(confidence) : x가 있을 때 y도 있는 비율(조건부 확률)

- association_rules 함수를 이용하여 향상도가 양의 상관관계에 있는 것 알아보기

```python
rules = association_rules(freq_item, metric = "lift", min_threshold = 1)
rules.sort_values('confidence', ascending = False, inplace = True)
rules
```
