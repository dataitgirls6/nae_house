## 신혼부부 인테리어 패턴 분석하기

### 데이터 불러오기

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cus =  pd.read_csv('housewarming2.csv')
prd = pd.read_csv('hou_prd_details.csv')
prd.drop(columns='Unnamed: 0', inplace=True)
```

```python
cus
```

```python
prd
```

### 데이터 조인

```python
master = pd.merge(cus, prd, on = 'id', how = 'inner')
master.info()
```

```python
master['작성일'] = pd.to_datetime(master['작성일'])
master.info()
```

### 분석에 사용할 칼럼 가져오기

```python
df = master[['id', '작성일', '제목', '가족형태', '공간', '평수구간', '스타일', '예산', '브랜드명', '상품명', '카테고리', '상품가격']]
df
```

### 신혼부부 구매 패턴 분석

```python
# 신혼부부 데이터만 가져오기
df = df[df['가족형태'].str.contains('신혼부부')]
df
```

```python
df['가족형태'].value_counts()
```

```python
# 신혼부부 총 유저 수
print(len(df['id'].unique()))
```

```python
# 유저별 몇 개의 상품을 구매하나?
customer_unique_transaction_series = df.groupby('id')['상품명'].nunique()
customer_unique_transaction_series.describe()
```

인테리어를 할 때 유저당 평균 55개의 상품 구입, 최소5개 - 최대191개

```python
# 유저별 아이템 카테고리 종류 개수를 탐색
customer_unique_item_series = df.groupby('id')['카테고리'].nunique()
customer_unique_item_series.describe()
```

인테리어를 할 때 유저당 평균 44개의 카테고리에서 구입, 최소3개 - 최대130개   

```python
# 가장 구매가 많은 카테고리 top 10 탐색
df['카테고리'].value_counts()
```

선물하기 좋은 제품 카페고리 선정: 가구(의자) > 데코(화분) > 조명 > 주방용품. 가구는 선물하기 어렵고 데코와 큰 차이가 나지 않음.   
가구는 제외하거나 소가구만 보는 것도 방법

```python
# 가장 구매가 많은 상품 top 10
item_count = df['상품명'].value_counts()[:10]
for idx, (val, cnt) in enumerate(item_count.iteritems(), 1):
    print("Top", idx, ":", val, cnt)
```

선물하기 좋은 상품 선정: 주방가전, 반려식물, 액자 위주의 선물기획

```python
# 공간별로 상품에 얼마를 지출하나?
df2 = df.dropna(subset=['공간','상품가격'])  # 상품가격이 null인 행 제거
df2.groupby('공간')['상품가격'].mean().sort_values(ascending = False)
```

아파트 > 단독주택 > 원룸&오피스텔 > 빌라&연립 > 사무공간 > 기타
- 아파트가 평균 지출액이 높은데 아파트에 거주하는 신혼부부가 많고 처음 입주할 때 전체 인테리어를 많이 하기 때문이라고 판단
