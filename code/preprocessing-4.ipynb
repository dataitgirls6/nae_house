## 평수 10평 미만 : 구간화

```python
import pandas as pd

df04 = pd.read_csv("df02.csv")
df04['평수구간'] = df04['평수구간'].apply(lambda x : x+5)
df04.head()
```

```python
def pyeongsu(x):
    if x == 12:
        return 5
    else:
        return x
```

```python
df04['평수구간'] = df04['평수구간'].apply(pyeongsu)
df04.head()
```

```python
df04.columns
```

```python
df04 = df04.drop('Unnamed: 0', axis=1)
df04.head()
```

```python
df04 = df04.reset_index()
df04.rename(columns = {'index':'id'}, inplace=True)
df04.head()
```

```python
df04.to_csv("housewarming.csv", index=False)
```
