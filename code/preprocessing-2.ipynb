### â— í‰ìˆ˜ ì¹´í…Œê³ ë¦¬í™”

```python
def pyeong(x):
    if x < 10:
        return 7
    elif 10 <= x < 20:
        return 10
    elif 20 <= x < 30:
        return 20
    elif 30 <= x < 40:
        return 30
    elif 40 <= x < 50:
        return 40
    elif 50 <= x < 60:
        return 50
    elif 60 <= x < 70:
        return 60
    else:
        return 70
```
